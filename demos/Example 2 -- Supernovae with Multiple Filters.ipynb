```python
values = np.linspace(0,10,100)
times = np.linspace(0,10,100)
object = 'linear'
time_total = 5 * 1000 #ms

# durations
duration_args = {
#     'time_min' : 50, #ms
#     'time_max' : 2000, #ms
    'time_total': time_total,  # ms
}

# frequencies
C4 = 261.6  # Hz
frequency_args = {
#     'frequency_min': C4 * 1. / 3.,
#     'frequency_max': C4 *1. / 2.,
    'frequency_min': C4 * 1. ,
    'frequency_max': C4 * 3,
#     'cents_per_value' : -680,
    'value_min': min(values),
    'value_max': max(values),
}

Tone = SonifyTool(values=values, times=times, frequency_args=frequency_args, duration_args=duration_args, length=.1)
# Tone = SonifyTool(values=flux, times=wavelength, frequency_args=frequency_args, duration_scale=(max(wavelength)-min(wavelength))/25000.)
Tone.SaveTone(filename='MultiTone_' + object +'.wav')
# Tone.PlotTone()
```
