## SNe

```python
fs1 = 12
colors = ['#DAA51B', '#38A6A5', '#2F8AC4', '#5F4690', '#DC143C']
test_alpha = 0

for SN in SNlist[0][:]:
    SN = SN[:-5]
    print(SN)
    
    # importing data mags and times

    SNAll = pd.read_json('./open_supernovae_catalog/' + SN + '.json')
    Phot = SNAll[SN]['photometry']
    mags, time, type_all = returnMagsTimesTypes(SNAll, SN, bandtype='B')

    values, times = mags, time
    
    SN_DF = pd.DataFrame({'values' : values, 'times' : times})
    SN_sorted = SN_DF.sort_values('times')
    
    maxtime = SN_sorted['times'][SN_sorted['values'].idxmax()]
    MaxFiltered = SN_sorted[SN_sorted['times'] < min(times)  + cut_off_time]
    
    values = np.array(MaxFiltered['values'])
    times = np.array(MaxFiltered['times'])
    
    # setting up frequency and duration arguments and generating tone

    # frequencies
    C4 = 261.6  # Hz
    frequency_args = {
        # 'frequency_min': C4, #* (1 / 2),
        'frequency_max': C4 * 3,
        'cents_per_value' : -680,
        'value_min': max(mags),
        'value_max': min(mags),
    }
    
    duration_args = {
        # 'time_min' : 50, #ms
        # 'time_max' : 2000, #ms
        'time_total': 1000,  # ms
    }
    
    # finding duration scale
    secscale = getScale(times=times, **duration_args)
    print(secscale)
    secscale = cut_off_time / 1000.  # MJD / ms
    print(secscale)
    
    allscales = np.array([1/3., 1/2.5, 1/2., 1/1.5, 1.]) * secscale # MJD / ms
    scale = allscales[0] # MJD / ms
    
    duration = cut_off_time/scale*1/1000 # s = day *ms/MJD *s/ms
    fps = n_images/duration #frames / (day *ms/MJD *s/ms)
    print('fps:',fps)
    print('duration:',duration)
    
# #     Tone = SonifyTool(values=values, times=times, frequency_args=frequency_args, duration_args=duration_args, length=.1)
    Tone = SonifyTool(values=values, times=times, frequency_args=frequency_args, duration_scale=scale)

#     filename = 'bestcases/' + str(SN) + '_' + str(cut_off_time/(scale*1000)) + 's.wav'
#     filename = 'bestcases/' + str(SN) + '_' + str(cut_off_time/(scale*1000)) + 's.mp3'
#     Tone.SaveTone(filename=filename)
#     filename = str(SN) + '_' + str(cut_off_time/(scale*1000)) + 's.mp3'
    filename = str(SN) + '_' + str(cut_off_time/(scale*1000)) + 's.wav'
    Tone.SaveTone(filename=filename, path='./plots_gifs/')
    
    # building and saving gif images
    
#     im_names = ["./plots_gifs/"+SN+"_%i.png" % int(i) for i in range(n_images)]
#     im_times = np.linspace(0,cut_off_time,n_images)    
    
    fig, ax = plt.subplots(figsize=(10,6))
#     fig.suptitle(SN)
    
    ax.invert_yaxis()
    
    def make_frame(t):
        ax.clear()
#         ax.scatter(times-min(times), values, s=1, marker='+', label=type_all)
        ax.scatter(times-min(times), values, s=500, marker=1, linewidth=1, color=colors[3])
        ax.scatter(times-min(times), values, s=15, marker='*', color=colors[-1], label='Observed Magnitudes')

        ax.set_xlabel(r'$\Delta MJD = MJD - MJD_{start}$  |  $MJD_{start} = $'+str(min(times)) + ' | days', fontsize=fs1)
        ax.set_ylabel(r'$m_B$', fontsize=fs1)

#         ax.legend(frameon=True, loc=1)
        ax.axvline(x=t*scale*1000, c='b') # s * MJD / ms * ms/s
        ax.set_xlim(0,cut_off_time)
        
        ax2 = ax.twinx()
        ax2.clear()
        ax2.semilogy(times-min(times), Tone.frequencies, alpha=test_alpha)
        ax2.set_ylabel('Frequency | Hz', fontsize=fs1)
        
        ax3 = ax.twiny()
        ax3.clear()
        ax3.plot(Tone.starttimes,values, alpha=test_alpha)
        ax3.plot(0,values[0], alpha=test_alpha)
        ax3.set_xlabel('Duration into Audio File | seconds', fontsize=fs1)

        ax.tick_params(which='major', direction='in')#, labelsize=fs2)
        ax.tick_params(which='minor', direction='in')#, labelsize=fs2)
        ax.tick_params(top=False, bottom=True, left=True, right=False, which='both')
        ax.minorticks_on()
        ax2.tick_params(which='major', direction='in')#, labelsize=fs2)
        ax2.tick_params(which='minor', direction='in')#, labelsize=fs2)
        ax2.tick_params(top=False, bottom=False, left=False, right=True, which='both')
        ax2.minorticks_on()
        ax3.tick_params(which='major', direction='in')#, labelsize=fs2)
        ax3.tick_params(which='minor', direction='in')#, labelsize=fs2)
        ax3.tick_params(top=True, bottom=False, left=False, right=False, which='both')
        ax3.minorticks_on()

        ax.legend(title=SN + ' | ' + type_all, loc=1)
        
        ax.text(0, max(values), '@Astro_Locke | @emsque')
        
        return mplfig_to_npimage(fig)

    animation = VideoClip(make_frame, duration=duration + 0.1)
    animation.write_videofile('./plots_gifs/'+SN+'.mp4', fps=fps)
```
